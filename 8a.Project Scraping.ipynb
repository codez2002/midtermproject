```python
#I try to populate different lists
#I am going through each row (tr) and look for different 
#cell elements (td) to upload the different lists

listheaders=[]
listRank=[]
listJob=[]
listSalary=[]
listNumberEd=[]
all_rows = stat_table.find_all('tr')
#count= counts rows
count=0

for r in all_rows:
    ranking = ""
    job = ""
    salary=""
    ed = ""
    col = 0
    if count > 0:
        all_td= r.find_all("td")
        for d in all_td:
            if col==0:
                ranking=d.get_text()
                listRank.append(ranking)
            
            elif col==1:
                job=d.get_text()
                listJob.append(job)
            
            elif col==2:
                salary=d.get_text()
                listSalary.append(salary)
            elif col==3:
                ed=d.get_text()
                listNumberEd.append(ed) 
            else:
                print("nada")
                
            col=col+1
        
    else:
        
        allheaders = r.find_all("th")
        col = 0
        for x in allheaders:
            listheaders.append(x.get_text())
        
    
    count=count+1
    
```
