# ATTRITION SAMPLE
I downloaded a dataset from Kaggle in order to find out different causes for employee attrition

```python
#I used panda to upload and read the spreadsheet in Jupyter and created a data frame.
#The spreadsheet has 35 columns and 1470 rows
import pandas as pd
```

```python
df=pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

print(type(df))
```

```python
df
```

```python
#I separate the populations, to see if there are any differences between people who left (Attrition =Yes)
#and people who are still at the company (Attrition =No)
dfattrition = df[df['Attrition'] == 'Yes']
dfNoattrition = df[df['Attrition'] == 'No']
```

```python
dfattrition 
```

```python
#I print both populations to make sure that the selection is correct
dfNoattrition
```

```python
#The chart below shows that there are significant differences in daily rates. So it is worthwile
#to investigate whether these differences could be a factor that make people leave

import matplotlib.pyplot as plt
ax = dfattrition [['Department','DailyRate']].sort_values('DailyRate').head(100).plot(kind='barh', title ="Attrition", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Department", fontsize=12)
ax.set_ylabel("DailyRate", fontsize=12)
#print(df["DailyRate"].describe()["mean"])

plt.show()
```

```python
#I decide to group the data by the median per Department so that I can visualise the differences between employees and leavers
#more clearly
#I had to calculate the numerical results too, to help me with the length of the x axis (not to small, not to large)
import numpy as np
grouped = dfattrition .groupby('Department')
grouped['DailyRate'].agg(np.median)
```

```python
import numpy as np
grouped = dfNoattrition .groupby('Department')
grouped['DailyRate'].agg(np.median)
```

```python
#I also want to know if one of the reasons for leaving could be that the workplace is too far.
import numpy as np
grouped = dfattrition.groupby('Department')
grouped['DistanceFromHome'].agg(np.median).plot.barh()

plt.xlim(0,20)
```

```python
#My next question is to find out whether there is any particular time and age at which the employee leave the company

import matplotlib.pyplot as plt
data = df[['YearsAtCompany','Age']].sort_values('Age')
plt.scatter( data.Age, data.YearsAtCompany,s=5)

ax.set_xlabel("YearsAtCompany", fontsize=12)
ax.set_ylabel("Age", fontsize=12)


plt.show()
```

```python
#On the chart below we can see that the majority of the leavers tend to go before they are 35 and roughly
#they spend 10 years or less with the company. Once employess have been 10 years with the company it will rare that they will leave
import matplotlib.pyplot as plt
data = dfattrition[['YearsAtCompany','Age']].sort_values('Age')
plt.scatter( data.Age, data.YearsAtCompany,s=5)

ax.set_xlabel("YearsAtCompany", fontsize=12)
ax.set_ylabel("Age", fontsize=12)
plt.grid(True)

plt.show()
```
